```python
#This program will determine the validity of a password according to stipulated criteria.
pas = input("Let's see if your password will work: ")
#After the program has the user's proposed password stored as the 'pas' variable it's subjected to the Tally function.
def Tally(pas):
    upper, lower, num, sym = 0, 0, 0, 0
    #This loop will categorize every character in the string.
    for i in range(len(pas)):
        if pas[i].isupper():
            upper += 1
        elif pas[i].islower():
            lower += 1
        elif pas[i].isdigit():
            num += 1
        else:
            sym +=1
    #This part of the function determines whether or not the password is of the appropriate length and has one of each category of character.
    if upper <= 1:
        print("Your password needs a capital letter.")
    elif lower <= 1:
        print("Your password needs a lowercase letter.")    
    elif num <= 1:
        print("Your passowrd needs a number.")
    elif sym <= 1:
        print("Your password needs a non-alphanumeric character.")
    elif len(pas) > 16 or len(pas) <6:
        print("The acceptable length of a password is 6 to 16 characters.")
    else:
        print("Well done! Here's your password: ", str(pas))
Tally(pas)
        
```

```python
#This script uses the sort function to arrange three numerical inputs for the list 'age' from lowest to highest.
age = []
per1 = input("What's the first person's age? ")
age.append(per1)
per2 = input("What's the second person's age? ")
age.append(per2)
per3 = input("What's the third person's age? ")
age.append(per3)
age.sort()
print("Here are the ages in ascending order: ", age)

```

```python
#This program identifies a particular candidate for admission as being eligible or ineligible according to the ratio of two inputs: the number of classes held and the number attended.
Classesheld = input("How many classes were there?")
Daysattended = input("How many of these classes were attended?")
Ratio = (int(Daysattended)/int(Classesheld))
print(Ratio)
if Ratio > .75:
    print("This is an acceptable ratio. This student is eligible.")
else:
    print("This is not an acceptable ratio. This student is not eligible.")
```
